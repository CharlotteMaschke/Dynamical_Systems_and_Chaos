```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def euler(fx, seedx,fy, seedy, dt, stop_change = 10e-5):
    """
    Run the euler method on a given formula

    Arguments:
    fx   -- a differential equation in form of a function f(x)
    seed -- the seed of this function x0
    dt   -- delta t as the step size for the euler method
    stop -- stopping criteria, value of value of change small enought to stop the function
            the function always calculates a minimum of 30 timesteps.
            default = 10e-5
    
    Output:
    solution -- values of X over given initial condition
    times    -- corresponding time steps
    
    """
    # initialize frame
    solutionx = []
    solutiony = []
    time = []
    
    # initialie first value
    x = seedx
    y = seedy
    t = 0
    
    # add first value to output
    solutionx.append(x)
    solutiony.append(y)
    time.append(t)
    
    
    while True:
        # calculate dy/dx
        changex = fx(x,y)
        changey = fy(x,y)
        
        t = t+dt
        # calculate new y
        x = x + (changex * dt)
        y = y + (changey * dt)
        # append solution
        time.append(t)
        solutionx.append(x)
        solutiony.append(y)
        # break if change is too small
        if changex < 10e-5:
            # add timestep as additional stopping condition ( min of t=40)
            if t>100:
                break
    
    return (solutionx, solutiony, time)
```

```python
def fy(x,y):
    y = -x+(1-x**2)*y
    return y

def fx(x,y):
    x = y
    return x
```

```python
# Run Euler's Method on given function with different initial conditions
solx1, soly1 ,time1 = euler(fx, 3, fy,-3, 0.05)
solx2, soly2 ,time2 = euler(fx, 0.2, fy,0.2, 0.05)
solx3, soly3 ,time3 = euler(fx, 5, fy,10, 0.05)
```
