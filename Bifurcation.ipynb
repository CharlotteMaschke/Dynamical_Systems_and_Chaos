```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def logistic_map(r, x):
    return r * x * (1 - x)
```

```python
def equation_A2(r, x):
    # plots the equation from homework Advanced 2
    return r * np.sin((np.pi*x)/2)
```

```python
def bifurcation(rs, ignore, iterations, x):
    fig = plt.Figure()    
    for i in range(0,iterations):
        x = logistic_map(rs,x)
        if ignore < i:
            plt.scatter(rs,x, c = 'black',marker='.',s= 0.001)
    return fig
```

```python
rs = np.arange(2.5,4,0.001)
iterations = 200
ignore = 5
seed = 0.2
```

```python
fig = bifurcation(rs, ignore, iterations, seed)
#plt.xlim(3.84, 3.86)
#plt.ylim(0.4, 0.6)
```
