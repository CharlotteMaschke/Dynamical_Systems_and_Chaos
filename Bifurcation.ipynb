```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def logistic_map(r, x):
    return r * x * (1 - x)
```

```python
def equation_A2(r, x):
    # plots the equation from homework Advanced 2
    return r * np.sin((np.pi*x)/2)
```

```python
def bifurcation(rs, ignore, toplot, iterations, x):
    fig = plt.Figure()    
    for i in range(0,iterations):
        x = logistic_map(rs,x)
        if ignore < i > ignore + toplot:
            plt.scatter(rs,x, c = 'black',marker='.',s= 0.05)
    return fig
```

```python
rs = np.arange(0,10,0.01)
iterations = 400
ignore = 100
toplot = 200
seed = 0.1
```

```python
fig = bifurcation(rs, ignore, toplot, iterations, seed)
#plt.xlim(3.3, 3.5)
#plt.ylim(0.2, 0.5)
```
