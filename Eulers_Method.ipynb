```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def euler(fx, seed, dt, stop_change = 10e-5):
    """
    Run the euler method on a given formula

    Arguments:
    fx   -- a differential equation in form of a function f(x)
    seed -- the seed of this function x0
    dt   -- delta t as the step size for the euler method
    stop -- stopping criteria, value of value of change small enought to stop the function
            the function always calculates a minimum of 30 timesteps.
            default = 10e-5
    
    Output:
    solution -- values of X over given initial condition
    times    -- corresponding time steps
    
    """
    # initialize frame
    solution = []
    time = []
    
    # initialie first value
    y = seed
    t = 0
    
    # add first value to output
    solution.append(y)
    time.append(t)
    
    
    while True:
        # calculate dy/dx
        change = fx(y)
        t = t+dt
        # calculate new y
        y = y + (change * dt)
        # append solution
        time.append(t)
        solution.append(y)
        # break if change is too small
        if change < 10e-5:
            # add timestep as additional stopping condition ( min of t=40)
            if t>30:
                break
    
    return (solution, time)
```

```python
def fx(T):
    y = 0.2*(20-T)
    return y
```

```python
# Run Euler's Method on given function with different initial conditions

# different delta t values
# function stops if change smaller than 0.1
sol05,time05 = euler(fx, 10, 0.5, 0.1)
sol1,time1 = euler(fx, 10, 1, 0.1)
sol2,time2 = euler(fx, 10, 2, 0.1)
```

```python
plt.plot(time2, sol2)
plt.plot(time1, sol1)
plt.plot(time05, sol05)


plt.legend(['delta t = 2','delta t = 1','delta t = 0.5'])
plt.xlabel("time [t]")
plt.ylabel("X(t)")
```

```python
#Real solution:
def real_solution(t, seed):
    y = 20-(20-seed)*(np.e**(-0.2*t))
    return y
```

```python
# Calculate Real Solution for seed 10
t = np.arange(1,31)
y_real = real_solution(t, 10)
```

```python
plt.plot(time2, sol2)
plt.plot(time1, sol1)
plt.plot(time05, sol05)
plt.plot(t, y_real)


plt.legend(['delta t = 2','delta t = 1','delta t = 0.5','REAL Solution'])
plt.xlabel("time [t]")
plt.ylabel("X(t)")
```
