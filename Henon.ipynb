```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def solution (fx, seedx,fy, seedy, iterations):
    """
    Outputs the time series plot of the logistic equation given some parameters

    Arguments:
    r            -- growth rate of population
    seed         -- the seed of this function x0
    iterations   -- number of iterations to calculate

    Output:
    out      -- values of X over time given initial condition
    
    """
    
    # initialize output and insert seed
    outx = []
    outx.append(seedx)
    
    outy = []
    outy.append(seedy)

    x = seedx
    y = seedy
    
    # iterate over iterations and calculate new y
    for i in range(iterations):
        # x as Population next year
        x_tmp = fx(x,y)
        y_tmp = fy(x,y)
        x = x_tmp
        y = y_tmp
        outx.append(x)
        outy.append(y)
        
    return(outx, outy)    
```

```python
def fy(x,y):
    b = 0.2
    y = b * x
    return y

def fx(x,y):
    a = 1.3
    x = y + 1 - (a*(x**2))
    return x
```

```python
# Run Solution on given function with different initial conditions
iters = 10000
solx, soly = solution (fx, 0.5,fy, 0.2, iters)
```

```python
# define how many iterations to skip for plotting
skip = 9000
```

```python
plt.figure()
plt.scatter(solx[skip:], soly[skip:])
plt.plot(solx[skip:], soly[skip:], alpha = 0.3)

plt.figure()
plt.plot(solx[skip:])

plt.figure()
plt.plot(soly[skip:])
```
